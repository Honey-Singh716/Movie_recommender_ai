```python
import ast
import pandas as pd
import pickle
```

```python
credits = pd.read_csv('../data/tmdb_5000_credits.csv')
```

```python
def parse_cast(cast_str):
    try:
        return ast.literal_eval(cast_str)
    except:
        return []
```

```python
data = pd.read_pickle('../models/cleaned_movie_data.pkl')
```

```python
data['cast_parsed'] = credits['cast'].apply(parse_cast)
```

```python
cast_rows = []

for _, row in data.iterrows():
    movie_id = row['id']
    movie_title = row['title']
    
    for cast in row['cast_parsed']:
        cast_rows.append({
            "cast_id": cast.get("cast_id"),
            "person_id": cast.get("id"),
            "name": cast.get("name"),
            "character": cast.get("character"),
            "order": cast.get("order"),
            "movie_id": movie_id,
            "movie_title": movie_title   
        })

cast_df = pd.DataFrame(cast_rows)
```
