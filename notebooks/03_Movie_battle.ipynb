```python
def get_movies(movie_name,data):
    movie = data[data["original_title"].str.lower() == movie_name.lower()]

    if movie.empty:
        return None

    movie = movie.sort_values("vote_count",ascending = False).iloc[0] 
    return movie   
```

```python
import pandas as pd

def movie_battle(movie1, movie2):
    comparison = {
        "IMDb Rating": [movie1['vote_average'], movie2['vote_average']],
        "Popularity": [movie1['popularity'], movie2['popularity']],
        "Runtime (min)": [movie1['runtime'], movie2['runtime']],
        "Budget": [movie1['budget'], movie2['budget']],
        "Revenue": [movie1['revenue'], movie2['revenue']],
        "Release Year": [
            int(movie1['release_date'][:4]),
            int(movie2['release_date'][:4])
        ]
    }

    return pd.DataFrame(
        comparison,
        index=[movie1['original_title'], movie2['original_title']]
    )
```

```python
m1 = get_movies("Inception", data)
m2 = get_movies("Interstellar", data)

if m1 is None or m2 is None:
    print("Movie not found")
else:
    battle_df = movie_battle(m1, m2)
    display(battle_df)
```
